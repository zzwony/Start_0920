````python
mpg['nclass'].unique()
````

````python
mpg.query('nclass == "suv"').assign(total = (mpg['cty'] + mpg['hwy'])/2).groupby('manufacturer').agg(mean_tot = ('total', 'mean')).sort_values('mean_tot',ascending=False).head(5)
````

## 데이터 합치기

````python
test1 = pd.DataFrame({'id' : [1,2,3,4,5],
                     'midterm' : [60, 80, 70, 90, 85]})
test2 = pd.DataFrame({'id' : [1,2,3,4,5],
                     'midterm' : [70, 83, 65, 95, 80]})
````

````python
#id 기준으로 합쳐서 total에 할당
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
````

````python
group_a = pd.DataFrame({'id':[1,2,3,4,5],
                       'test': [60,80,70,90,85]})

group_b = pd.DataFrame({'id':[6,7,8,9,10],
                       'test': [70, 83, 65, 95, 80]})
````

````python
group_all = pd.concat([group_a, group_b])
group_all
````
