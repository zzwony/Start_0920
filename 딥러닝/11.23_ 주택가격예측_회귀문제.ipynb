# 주택 가격 예측: 회귀 문제

````python
# 데이터 로드
from keras.datasets import boston_housing

(train_data, train_targets), (test_data, test_targets) = (boston_housing.load_data())
````

````python
train_data.shape
# 카럼의 개수가 13개
````

````python
train_targets.shape
````

````python
train_targets[:4]
# 가격
````

````python
# 데이터 정규화(전처리)
# 정규화: 0과 1사이의 값으로 만들어주는것
# 같은 칼럼(속성)별로 평균을 구해줘야한다. 늘 세로로 해줘야한다.
````

````python
# 평균
mean = train_data.mean(axis=0)
train_data -= mean  
# -= 왼쪽 변수에 오른쪽 값을 빼고 그 결과를 왼쪽 변수에 할당

# 표준편차
std = train_data.std(axis=0)
train_data /= std
# /= 왼쪽 변수에 오른쪽 값을 나누고 그 결과를 왼쪽 변수에 할당

test_data -= mean
test_data /= std
````

````python
# 모델 만들기
model = Sequential([
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)
])

model.compile(optimizer='rmsprop',
             loss='mse',
             metrics='mae')
````

````python
history = model.fit(train_data, train_targets,
         epochs=100, verbose=0, batch_size=16)
````

````python
history.history['mae']
````

````python
plt.plot(history.history['mae'])
````

````python
preds = model.predict(test_data)
preds
````

````python
model.evaluate(test_data, test_targets)
````

## k-fold Cross Validation

````python
k = 4
num_val_samples = len(train_data) // k
num_epochs = 100
all_scores = []

for i in range(k):
    print(f'{i} 번째 폴드 처리중')
    # 검증 데이터 준비
    val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = train_targets[i * num_val_samples: (i + 1) * num_val_samples]
    
    # 훈련 데이터 준비
    partial_train_data = np.concatenate(
        [train_data[:i * num_val_samples],
        train_data[(i + 1) * num_val_samples:]], axis=0)
    partial_train_targets = np.concatenate(
        [train_targets[:i * num_val_samples],
        train_targets[(i + 1) * num_val_samples:]], axis=0)
   
    # 모델 구성
    model = Sequential([
        Dense(64, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1)
    ])
    
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    
    model.fit(partial_train_data, partial_train_targets,
             epochs=num_epochs, batch_size=16, verbose=0)
    val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=0)  # 검증 세트로 모델 평가
    all_scores.append(val_mae)
````

````python
all_scores
````

````python
np.mean(all_scores)
````

이 결과는 평균적으로 26000달러 정도가 차이가 난다.

주택 가격의 범위가 1만 달러에서 5만 달러 사이인 것을 감안하면 비교적 큰 값이다.

그래서 훈련 반복 횟수를 더 늘려보기로 했다.

````python
# 에포크를 500으로 늘리고 개선 정도를 알기 위해
# 각 에포크마다 각 폴드의 검증 점수를 저장하기

num_epochs = 500
all_mae_histories = []

for i in range(k):
    print(f'{i} 번째 폴드 처리중')
    # 검증 데이터 준비
    val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = train_targets[i * num_val_samples: (i + 1) * num_val_samples]
    
    # 훈련 데이터 준비
    partial_train_data = np.concatenate(
        [train_data[:i * num_val_samples],
        train_data[(i + 1) * num_val_samples:]], axis=0)
    partial_train_targets = np.concatenate(
        [train_targets[:i * num_val_samples],
        train_targets[(i + 1) * num_val_samples:]], axis=0)
   
    # 모델 구성
    model = Sequential([
        Dense(64, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1)
    ])
    
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    
    history = model.fit(partial_train_data, partial_train_targets,
                        validation_data = (val_data, val_targets), 
                        epochs=num_epochs, batch_size=16, verbose=0)
    mae_history = history.history['val_mae']
    all_mae_histories.append(mae_history)
````

````python
# 모든 폴드에 대해 에포크의 MAE 점수 평균 구하기
average_mae_history = [
    np.mean([x[i] for x in all_mae_histories]) for i in range(num_epochs)]
````

````python
plt.plot(range(1, len(average_mae_history) + 1), average_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
````

출력된 그래프는 범위가 크기 때문에 보기 어렵다.

처음 몇 번의 에포크 동안 MAE가 그 이후 에포크가 너무 높다.

그러므로 나머지 곡선 부분과 스케일이 크게 다른 처음 10개의 데이터 포인트를 제외시킨 후 그래프 다시 그려보기

````python
# 처음 10개의 데이터 포인트를 제외한 검증 점수 그래프 그리기
truncated_mae_history = average_mae_history[10:]
plt.plot(range(1, len(truncated_mae_history) + 1), truncated_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
````
