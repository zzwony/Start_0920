```python
from transformers import TFBertForMaskedLM

model_checkpoint = "klue/bert-base"
model = TFBertForMaskedLM.from_pretrained(model_checkpoint, from_pt=True)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
df_kpop_image_texts = [['이 밤 그날의 [MASK]을 당신의 창 가까이 보낼게요', '모래 위에 적힌 [MASK]처럼','나 우리의 첫 [MASK]을 떠올려', '나의 [MASK] 안에 모든 말을 다 꺼내어 줄 순 없지만'],  # 첫번째 노래
                       ['어제의 너는 [MASK]을 타고 멀리 후회도 없이 미련 없이 날아가', '[MASK]에 비친 너의 모습이 슬퍼','저 파란 [MASK] 눈부신 별들도 슬프게 내린 비는 꽃을 피우고', '너의 눈물은 [MASK] 사랑을 피울 거야'], # 두번째 노래 
                       ['눈 앞에선 우리의 거친 길은 알 수 없는 [MASK]와 벽', '멈춰져 버린 이 [MASK]','그대의 부드러운 [MASK]이 이 순간 따스하게 감겨오네', '희미한 [MASK]을 난 쫓아가'], # 세번째 노래
                       ['울었던 웃었던 소년과 [MASK]가 그리워 나', '이듬해 질 녘 [MASK] 피는 봄','한여름 밤의 [MASK]', '비스듬히 씩 비웃듯 [MASK] 무늬의 무지개'], # 네번째 노래
                       ['밤하늘의 [MASK]이 빛난 것처럼', '꽃이 피고 [MASK] 날이 와도','그저 바라보는 [MASK] 그 하나로', '나는 [MASK]라면 기다릴 수 있을 텐데'],
                       ['너희 [MASK]을 보고 있어도 보고 싶다', '허공을 떠도는 작은 [MASK]처럼, 작은 먼지처럼','[MASK]이 떨어져요 또 조금씩 멀어져요', '벚꽃이 피나봐요 이 [MASK]도 끝이 나요']] # 다섯번째 노래
```
