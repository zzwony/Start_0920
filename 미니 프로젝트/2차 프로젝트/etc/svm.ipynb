```python
from sklearn.model_selection import train_test_split
import pandas as pd

path = '/content/drive/My Drive/Colab Notebooks/sample/lmemdata/'

user_clr_df = pd.read_csv(path+'lmembers_features_통합(train또는1~6분기데이터).csv')
user_clr_df[['거주지역','구매금액이큰제휴사','방문빈도가높은제휴사']] = user_clr_df[['거주지역','구매금액이큰제휴사','방문빈도가높은제휴사']].astype('category')

test = pd.read_csv(path+'lmembers_features_통합(test또는2~7분기데이터).csv')
test[['거주지역','구매금액이큰제휴사','방문빈도가높은제휴사']] = test[['거주지역','구매금액이큰제휴사','방문빈도가높은제휴사']].astype('category')

user_clr_df = user_clr_df.set_index('고객번호')
test = test.set_index('고객번호')
```

```python
# 어떤  feature를 사용할 것인가 결정하는 부분 , 아래 PCA 쓸거면 실행 안해도 됨
user_clr_df = user_clr_df[['A_Monetary_실제지출', 'B_Monetary_실제지출', 'C_Monetary_실제지출','D_Monetary_실제지출', 'Monetary','fluctuation', '분기매출증가']]
test = test[['A_Monetary_실제지출', 'B_Monetary_실제지출', 'C_Monetary_실제지출','D_Monetary_실제지출', 'Monetary','fluctuation', '분기매출증가']]
```

```python
# PCA 수행
# Feature가 의미있는지 판단해보기
# 실제로 적용할지 안할지는 생각해봐야함

from sklearn.decomposition import PCA

pca_dimension = 6

pca = PCA(n_components=pca_dimension)   # 몇차원으로 압축할 것인가
pca_transformed = pca.fit_transform(user_clr_df)
pca_transformed = pd.DataFrame(pca_transformed)
pca.explained_variance_ratio_
```

```python

X_label = user_clr_df.분기매출증가
X_features = user_clr_df.drop(columns =['분기매출증가'])
X_features = pd.get_dummies(X_features)

X_label_t = test.분기매출증가
X_features_t = test.drop(columns =['분기매출증가'])
X_features_t = pd.get_dummies(X_features_t)

# 전체 데이터 중 80%는 학습용 데이터, 20%는 테스트용 데이터 추출
X_train, X_val, y_train, y_val=train_test_split(X_features, X_label,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier

model = LogisticRegression()
svm_model = SGDClassifier(alpha = 0.1, average = False, class_weight = None, early_stopping = False, epsilon = 0.1, n_jobs = -1, l1_ratio = 0.8900000001, eta0 = 0.001,max_iter = 1000 , fit_intercept = False, learning_rate = 'optimal', loss = 'hinge', random_state = 3)
model.fit(X_train, y_train)
svm_model.fit(X_features, X_label)
print('lr', model.score(X_train, y_train))
print('svm', svm_model.score(X_features, X_label))
# test
print('lr', model.score(X_features_t, X_label_t))
print('svm', svm_model.score(X_features_t, X_label_t))
```
