####0번 군집 특징 정리와 솔루션 방향
< 0번 군집의 특징 >
1. 일반, 편의, 전문 상품이 구매 감소추세를 보이고 있다.
2. B 제휴사 구매 감소추세를 보임
3. A 제휴사에 지출이 쏠려있다.
- 1~3에 따라서 일반, 편의, 전문 상품을 A 와 B 중에 어디서 주로 구매하는지 확인하여 솔루션을 낸다.
- 만약 A 제휴사에서 해당 분류의 상품을 주로 구매한다면, 비슷한 고객의 데이터를 기반으로 상품추천을 해주고
- B 제휴사에서 해당 분류의 상품을 주로 구매한다면, B 제휴사 이용 자체가 줄어드는 것이므로 이에 맞는 솔루션이 필요하다. 
  - (솔루션: B 제휴사의 온라인 판매를 타겟으로? 하려 하였으나 뭐가 온라인인지 모른다. 온라인이 들어간 중분류명, 중분류코드로 산출한 데이터 외에는 모른다. 그러나 B 제휴사 이용횟수에 비해 B 제휴사 온라인 중분류의 매출액이 현저히 차이가 나서, 이 데이터는 무조건 구멍이 있다.)

< 추가 체크 사항 >

3. B 제휴사의 온라인 구매가 줄어서 그런 것인지 데이터로 확인.

< 확인결과>
- 결론 1 : 0번 군집은 A제휴사에 비해서 B제휴사가 일반/비내구/편의 상품 구매횟수가 높고, B에 비해 A제휴사의 전문 상품 구매횟수가 높다. 따라서 B제휴사 구매 감소추세와 일반, 편의 상품 구매 감소추세로 미루어보아 생활용품 구매에 있어 L사 이용이 줄어들고 있고,

=> A제휴사에서 구매를 잘 하지 않는다.

3. 전문 상품(명품)은 A 제휴사에 있다.
결론 : 1~3에 따라 A 제휴사의 명품 구매를 유도하는 솔루션을 낼 수 있다.
