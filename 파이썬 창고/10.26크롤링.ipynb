```python
import urllib.request
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%9A%B0%ED%81%AC%EB%9D%BC%EC%9D%B4%EB%82%98"
```

```python
from urllib import request # 소스코드를 받아오기 위함
```

```python
html = request.urlopen(url).read()
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
# find_all() 다 찾겠다라는 함수
soup.find_all(class_ = "news_tit")
# find만 쓰면 딱 한개만 보여준다.
```

```python
title = soup.find_all(class_ = "news_tit")
title
```

```python
len(title)
```

```python
# 원하는 기사로 뽑기
title[0].text  # 안에 있는 글자만 쏙 뽑아내었다.
```

```python
[tit.text for tit in title]
```

```python
# 이렇게 하면 링크가 뜨는데 링크만 추출하려면?
title[0]
```

```python
title[0].attrs['href']
# 속성의 이름이 href이다. 라는 뜻
```
