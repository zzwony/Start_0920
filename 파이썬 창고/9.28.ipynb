# DataFrame의 칼럼 사용하기

````python
frame = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1),
                      'c': ['one', 'one', 'one', 'two', 'two',
                            'two', 'two'],
                      'd': [0, 1, 2, 0, 1, 2, 3]})
frame
````

````python
frame = frame.set_index('c')  #c라는 칼럼이 사라지고 인덱스로 변함
frame
````

````python
frame = frame.reset_index()  #처음으로 되돌리려고 하기 때문에 괄호에 안써줘도 된다.
frame
````

# 데이터 합치기

## 두 데이터프레임 합치기

````python
df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],
                   'data2':range(3)})
````

````python
df1
````

````python
df2
````

````python
pd.merge(df1, df2)
````

````python
pd.merge(df1, df2, how = 'inner')
````

````python
pd.merge(df1, df2, how = 'outer')  # 아까 없었던 c,d부분이 들어왔다.
````

````python
pd.merge(df1, df2, how = 'outer', indicator = True)  #어디서 날라왔는지 알려준다.
````

````python
pd.merge(df1, df2, how = 'left')  #왼쪽꺼는 다 살리라는 뜻
````

````python
pd.merge(df1, df2, how = 'right')
````

````python
pd.merge(df1, df2, on = 'key', how = 'outer')
````

````python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)}) 
````

````python
df3
````

````python
df4
````

````python
317
# 공통된건 없지만 내용상으로 같을 때 합치기
pd.merge(df3, df4, left_on = 'key', right_on = 'rkey')
````

````python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]}) 
````

````python
left
````

````python
pd.merge(left, right, on='key1')
````
