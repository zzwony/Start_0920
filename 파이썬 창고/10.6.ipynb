```python
years = range(1880, 2011)
columns = ['name','gender','births']
df = pd.DataFrame()

for year in years:
    frame = pd.read_csv(f'yob{year}.txt', names = columns)
    frame['year'] = year
    df = pd.concat([df, frame])
    
df
```

```python
df.set_index('name')
```

```python
names = df.set_index('name').reset_index()
names[:4]
```

```python
total_births = names.pivot_table(values='births', index='year', columns='gender', aggfunc=sum)
total_births.head()
```

```python
total_births.plot(kind='line')
```
