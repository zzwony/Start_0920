````python
raw_welfare = pd.read_spss('Koweps_hpwc14_2019_beta3.sav')
raw_welfare
````

````python
welfare = welfare.rename(
    columns = {'h14_g3':'sex', 'h14_g4':'birth', 'h14_g10':'marriage_type',
              'h14_g11':'religion', 'p1402_8aq1':'income',
              'h14_eco9':'code_job', 'h14_reg7':'code_region'})
````

````python
welfare[:2]
````

````python
welfare['sex'].value_counts()
````

````python
# sex 별로 income 평균
welfare.groupby('sex')['income'].mean()
````

````python
#sex에서 1은 male, 2는 female로 바꾸기
welfare['sex'] = np.where(welfare['sex']==1,'male','female')
welfare[:4]
````

````python
welfare.groupby('sex')[['income']].mean()
````

````python
df = welfare.groupby('sex', as_index=False)[['income']].mean()
df
````

````python
sns.barplot(data = df, x = 'sex', y = 'income')
````

- 나이별로 월급 평균 구하기

````python
welfare[:2]
````

````python
min(welfare.birth)
max(welfare.birth)
````

```
2018년이 1살
2017년이 2살
2016년이 3살
...
age 열을 만들어라.
```

````python
welfare['age'] = 2019 - welfare['birth']
welfare[:2]
````

````python
# age 열을 int로 바꾸기
welfare['age'] = welfare['age'].astype('int')
welfare.info()
````

````python
#나이별로 소득의 평균을 구해주세요.
df = welfare.groupby('age').agg('mean')[['income']]
````

````python
df = df.reset_index()
df
````

````python
#df로 lineplot을 그려보기
sns.lineplot(df, x = 'age', y= 'income')
````

```
새 칼럼 : age_gen
30이하 : young
60이하 : middle
나머지 : old
```

````python
welfare['age_gen'] = np.where(welfare['age'] <= 30, 
                              'young', np.where(welfare['age'] <= 60, 
                                                'middle', 'old')) 
welfare
````

````python
#'category'로 데이터타입을 바꿔 주세요.

welfare['age_gen'] = welfare['age_gen'].astype('category')
welfare.info()
````

````python
#세대별 소득평균, income, age_gen, gender

df_gen = welfare.groupby(['age_gen','sex']).agg('mean')[['income']]
df_gen
````

````python
df_gen = df_gen.reset_index()
````

````python
#세대별 성별 소득평균

sns.barplot(data = df_gen, x = 'age_gen', y = 'income', hue = 'sex', 
            order = ['young','middle','old'], palette = 'Set1')
````
